```python
pattern = re.compile('\D')
for index, row in list_tables.iterrows():
    table_id = '{}.{}'.format('uhma', row['Tables_in_uhma'])
    print('Loading Table {} ...'.format(table_id))
    df = pd_mod.read_sql('SELECT * FROM '+ str(table_id), conn)
    
    #set correct format for names and delete all non numerical values for phones
    if row['Tables_in_uhma']=='usuario':
        df['correo']=df['correo'].str.lower()
        df['nombre']=df['nombre'].str.title()
        df['nombre_usuario']=df['nombre_usuario'].str.title()
        df['apellido_paterno']=df['apellido_paterno'].str.title()
        df['apellido_materno']=df['apellido_materno'].str.title()
        df['telefono'].replace(pattern, '')
        df['celular'].replace(pattern, '')
        
    # transform date from unix timestamp dates 
    elif row['Tables_in_uhma'] in list_tables_with_dates:
        df['fecha']= pd.to_datetime(df['fecha'].to_numpy(), unit='ms')
        

    else:
        df
        
   
    #export table to BQ
    pd_gbq.to_gbq(df, table_id,
              project_id=PROJECT_ID,
              if_exists='replace',
              chunksize=10000000,
              progress_bar=True)
    print('Exporting tables csv...')
    if not os.path.exists(DIRECTORY):
        os.makedirs(DIRECTORY)

    df.to_csv('{}/{}.csv'.format(DIRECTORY, row['Tables_in_uhma']),
      index=None,
      header=True)
```
